```python
import csv
postits = []
with open('Mural-HT-1616167508796.csv', newline='',encoding='utf-8') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        postits.append(' '.join(row))
postits = postits[1:]
postits_text = [postitlinha.split(",")[1].replace("\"",'').lower() for postitlinha in postits]
postits_text_votados = [postitlinha.split(",")[1].replace("\"",'').lower() for postitlinha in postits if postitlinha.split(",")[9].replace("\"",'').lower() not in ('0','')]
postits_text_votos  = [postitlinha.split(",")[9].replace("\"",'').lower() for postitlinha in postits if postitlinha.split(",")[9].replace("\"",'').lower() not in ('0','')]
```

```python
classes_gramaticais = ["NOUN","VERB"]
nv = Nuvem()
nv.set_classes_gramaticais(classes_gramaticais)
nv.set_documento(postits_text_votados)
nv.set_votos(postits_text_votos)
nv.show_jupiter()
```
